```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales20_df = pd.read_csv("resources/athletic_sales_2020.csv")
athletic_sales21_df = pd.read_csv("resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
display(athletic_sales20_df)
```

```python
# Display the 2021 sales DataFrame
display(athletic_sales21_df)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales20_df.info()
```

```python
# Check the 2021 sales data types.
athletic_sales21_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_combined_df = pd.concat([athletic_sales20_df, athletic_sales21_df], axis="rows", join="inner", ignore_index=True)
athletic_sales_combined_df
```

```python
# Check if any values are null.
athletic_sales_combined_df.isnull().sum()
```

```python
# Check the data type of each column
athletic_sales_combined_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_combined_df["invoice_date"] = athletic_sales_combined_df["invoice_date"].astype("datetime64[ns]")
athletic_sales_combined_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales_combined_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_sales = athletic_sales_combined_df.groupby(['region', 'state', 'city']).agg(Total_Products_Sold=("units_sold", "sum"))

# Show the top 5 results.
grouped_sales = grouped_sales.sort_values("Total_Products_Sold", ascending=False)
grouped_sales.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_sales = athletic_sales_combined_df.pivot_table(
    index=["region", "state", "city"],
    values="units_sold",
    aggfunc="sum"
)

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_sales = pivot_sales.rename(
    columns={"units_sold": "Total_Products_Sold"}
)

# Show the top 5 results.
pivot_sales = pivot_sales.sort_values("Total_Products_Sold", ascending=False)
pivot_sales.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
region_salesg_df = athletic_sales_combined_df.groupby(['region', 'state', 'city']).agg({'total_sales': 'sum'}).rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
region_salesg_df = region_salesg_df.sort_values(by='Total Sales', ascending=False)
region_salesg_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
region_salesp_df = athletic_sales_combined_df.pivot_table(index=['region', 'state', 'city'], values='total_sales', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
region_salesp_df = region_salesp_df.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
region_salesp_df = region_salesp_df.sort_values(by='Total Sales', ascending=False)

region_salesp_df.head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailer_salesg_df = athletic_sales_combined_df.groupby(['retailer', 'region', 'state', 'city']).agg({'total_sales': 'sum'}).rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
retailer_salesg_df = retailer_salesg_df.sort_values(by='Total Sales', ascending=False)
retailer_salesg_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_salesp_df = athletic_sales_combined_df.pivot_table(index=['retailer', 'region', 'state', 'city'], values='total_sales', aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
retailer_salesp_df = retailer_salesp_df.rename(columns={'total_sales': 'Total Sales'})

# Show the top 5 results.
retailer_salesp_df = retailer_salesp_df.sort_values(by='Total Sales', ascending=False)

retailer_salesp_df.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
athletic_sales_combined_df["invoice_date"] = athletic_sales_combined_df["invoice_date"].astype("datetime64[ns]")
Womens_filter_df = athletic_sales_combined_df[athletic_sales_combined_df["product"] == "Women's Athletic Footwear"]
Womens_filter_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_salesg_df = Womens_filter_df.groupby(['retailer', 'region', 'state', 'city']).agg(Womens_Footwear_Units_Sold=("units_sold", "sum"))

# Show the top 5 results.
womens_salesg_df = womens_salesg_df.sort_values("Womens_Footwear_Units_Sold", ascending=False)
womens_salesg_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_salesp_df = Womens_filter_df.pivot_table(
    index=["retailer", "region", "state", "city"],
    values="units_sold",
    aggfunc="sum"
)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_salesp_df = womens_salesp_df.rename(
    columns={"units_sold": "Womens_Footwear_Units_Sold"}
)

# Show the top 5 results.
womens_salesp_df = womens_salesp_df.sort_values("Womens_Footwear_Units_Sold", ascending=False)
womens_salesp_df.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_salesd_df = Womens_filter_df.pivot_table(index = 'invoice_date', values = 'total_sales')
womens_salesd_df['total_sales'] = womens_salesd_df['total_sales'].astype(int)
# Optional: Rename the "total_sales" column to "Total Sales"
womens_salesd_df = womens_salesd_df.rename(columns = {'total_sales':'Total Sales'})

# Show the table.
womens_salesd_df
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = womens_salesd_df.resample('D', on='date').agg({'Total_Sales': 'sum'}).reset_index()

# Sort the resampled pivot table in descending order on "Total Sales".
sorted_daily_sales = daily_sales.sort_values(by='Total_Sales', ascending=False)
sorted_daily_sales
```
